```python
import pandas as pd
import numpy as np
```

```python
like_raw=pd.read_csv('likes.csv')

like_raw
post=like_raw['post'].drop_duplicates().reset_index(drop=True)
account=like_raw['account'].drop_duplicates().reset_index(drop=True)
like=pd.DataFrame(index=account,columns=post)
like[:]=0

print(type(like))
for i in range(len(account)):
    ac=like_raw[like_raw['account']==account[i]].reset_index(drop=True)
    
    for j in range(len(ac)):
        like[ac['post'][j]][account[i]]=1
        
    print("clear",i)
    
    
```

```python
like.to_csv('like_p.csv',sep=',',index=False)
```
